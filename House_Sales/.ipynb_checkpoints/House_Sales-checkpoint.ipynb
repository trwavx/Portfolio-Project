```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = df2.drop(columns = ['price'])
```

```python
y = df2.price
```

```python
print('x size: ', x.shape)
print('y size: ', y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
print('x train size: ', x_train.shape)
print('y train size: ', y_train.shape)
```

```python
print('x test size: ', x_test.shape)
print('y test size: ', y_test.shape)
```

# Linear Regression

```python
lr = LinearRegression()

lr.fit(x_train, y_train)
print('Score: ', lr.score(x_test, y_test))
```

```python
yhat = lr.predict(x_test)

ax1 = sns.distplot(df['price'], hist = False, color = "r", label = "Actual Value")
sns.distplot(yhat, hist = False, color = "b", label = "Predict Values", ax = ax1)

plt.title('Actual price vs Predict price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Features')
plt.show()
```

We can see that the fitted values are reasonably close to the actual values since the two distributions overlap a bit. However, there is definitely some room for improvement.

# Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
Input = [('scaler', StandardScaler()), ('poly', PolynomialFeatures(include_bias=False)), ('linear', LinearRegression())]
```

```python
pipe = Pipeline(Input)

pipe.fit(x_train, y_train)
print('Score: ', pipe.score(x_test, y_test))
```

```python
yhat2 = pipe.predict(x_test)

ax1 = sns.distplot(df.price, hist=False, color='r', label='Actual Price')
sns.distplot(yhat2, hist=False, color='b', label='Fitted Price', ax=ax1)

plt.title('Polynomial Actual price vs Fitted price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Features')
plt.show()
```

We can see that the fitted values way close to the actual value. But we have a little overfitting.

We will now use Ridge Regression to regularize and reduce the standard errors to fix the over-fitting in this model.

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
r = Ridge(alpha=0.1)

r.fit(x_train, y_train)
r.score(x_test, y_test)
```

```python
yhat3 = r.predict(x_test)

ax1 = sns.distplot(df.price, hist=False, color='r', label='Actual Price')
sns.distplot(yhat3, hist=False, color='b', label='Predict Price', ax=ax1)

plt.title('RR Actual price vs Fitted price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Features')
plt.show()
```

We can see that the fitted values are reasonably close to the actual values since the two distributions overlap a bit. However, there is definitely some room for improvement.
